```python
maps = read_maps("/home/gonzalo/sw/git/ICAROS/map_7707.h5")
corr = apply_all_correction(maps, apply_temp=False, norm_strat = norm_strategy.kr)

data = load_dsts(glob.glob("/home/gonzalo/data/NEXT/7707/*.h5"), "DST", "Events")
```

```python
ecorr = data.S2e * corr(data.X, data.Y, data.Z, data.time) * 1e3
```

```python
sel  = in_range(data.X.values, -40, -32)
sel &= in_range(data.Y.values, 120, 128)
plt.hist(ecorr[sel], np.linspace(0, 100, 101));
```
